```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/gym.csv'
gym = pd.read_csv(url)
```

```python
fig = px.box(gym, x = 'sex', y = 'weight', notched = True)
fig.show()
```

```python
gym = gym[gym['weight'] <104]
gym.head()
```
