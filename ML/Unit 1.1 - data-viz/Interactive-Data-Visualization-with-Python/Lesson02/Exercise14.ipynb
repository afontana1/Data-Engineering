# Creating a hexagonal binning plot

```python
import seaborn as sns
```

```python
mpg_df = sns.load_dataset("mpg")
```

```python
## set the plot style to include ticks on the axes.  
sns.set(style="ticks")
```

```python
## hexbin plot
sns.jointplot(mpg_df.weight, mpg_df.mpg, kind="hex", color="#4CB391")
```
