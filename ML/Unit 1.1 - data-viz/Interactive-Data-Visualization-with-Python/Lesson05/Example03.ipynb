```python
import numpy as np
from bokeh.plotting import output_notebook, figure, show
```

```python
output_notebook()
```

```python
#prepare some data
x = np.arange(5)
y = [6, 7, 2, 4, 5]
```

```python
# create a new plot specifying plot_height, plot_width, with a title and axis labels.
p = figure(plot_height=300, plot_width=700,title="simple line example",
x_axis_label='x', y_axis_label='y')
```

```python
# add a line renderer with legend and line thickness
p.line(x, y, legend="Temp", line_width=3)
```

```python
# show the results
show(p)
```
