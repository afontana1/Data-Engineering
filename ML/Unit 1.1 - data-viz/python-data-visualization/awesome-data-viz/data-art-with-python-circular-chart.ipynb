## Library

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
```

## Circular Chart

**Prior knowledge**

1. Cartesian and Polar Coordinate Systems (Of course, polar charts can also be used.)
    - Sin, Cos function
    - Radian <-> Degree
2. Modular Operation for Patterns


## Parameters

- Period 
- Cycle 
- Depth 

```python
period = 15
cycle = 4
depth = 20
degree = 2 * 3.141592 / (cycle*period)
total_size = period * cycle * depth

# colormap
cmap = sns.color_palette("cubehelix", period)
cmap = cmap * (total_size // 15)

# ETC
distance_from_center = 0.2
rotation_factor = 3
size_factor = 2.4
```

## Plot

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7), dpi=250)

points = np.arange(total_size)

dist = (points // (cycle*period) + 1) / depth + distance_from_center
x = np.sin(points * degree + dist * rotation_factor ) * dist 
y = np.cos(points * degree + dist * rotation_factor ) * dist 
s = ((points % period + 1) * dist / 3) ** size_factor

ax.scatter(x, y, s=s, color=cmap)

# Plot Settings
for spine in ['top', 'bottom', 'left', 'right']:
    ax.spines[spine].set_visible(False)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
