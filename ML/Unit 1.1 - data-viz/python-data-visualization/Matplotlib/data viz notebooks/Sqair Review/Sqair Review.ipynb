# Squair Air Purifier Review

Winter is coming, and we know what that means in Ulaanbaatar. Extremely high levels of air pollution. Even with the recent [ban on the burning of raw coal](https://montsame.mn/en/read/189503) it is possible (even likely) that levels of the particularly dangerous pollutant PM2.5 will remain high during the winter months.

During these months a pollution mask and air purifier are essential equipment to protect yourself and your family. Air purifiers can cost anywhere from 150USD-500USD or more. This means many families choose to forgo air purifiers completely, at the cost of their entire families health. Smart Air, a social enterprise founded in 2013, aims to change this situation with the launch of their new no-nonsense air purifier, the Sqair. Funded through a [Kickstarter campaign](https://www.kickstarter.com/projects/smartair/the-sqair-the-worlds-most-cost-effective-air-purifier/description) in June, the Sqair aims to be the world's most cost effective air purifier. Priced at a very competitive â‚®320,000, this compact air purifier claims to be as effective as larger, more expensive units. 

Sqair was kind enough to lend me a unit for a few days to test it out. I wanted to test a few things that I've found to be key points for consumers interested in buying an air purifier.
- Cost - Does it seem to be a good value for the product? This is less loaded than asking, "How much is your health worth?"
- Appearance - Does it look good a in a variety of settings?
- Usability - How easy is it to use? 
- Efficacy - Does it actually clean the air? How fast?

Each of these categories was easy to evaluate except for efficacy. For that I had to conduct my own experiment in a controlled environment to measure how quickly PM2.5 was cleaned from the air. See below for the results!

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('G:\My Drive\Data Science\mds.mplstyle')
```

```python
df = pd.read_csv('./data/experiment-data.csv')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df['Experiment'].value_counts()
```

```python
df['Experiment'].unique()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Minute'][df['Experiment'] == 'Squair off'], df['PM2.5 Value'][df['Experiment'] == 'Squair off'] , linewidth=2)
ax.plot(df['Minute'][df['Experiment'] == 'Squair - 1'], df['PM2.5 Value'][df['Experiment'] == 'Squair - 1'] , linewidth=2)
ax.plot(df['Minute'][df['Experiment'] == 'Squair - 2'], df['PM2.5 Value'][df['Experiment'] == 'Squair - 2'] , linewidth=2)
ax.plot(df['Minute'][df['Experiment'] == 'Squair - 3'], df['PM2.5 Value'][df['Experiment'] == 'Squair - 3'] , linewidth=2)
ax.set_xlabel('Minute')
ax.set_ylabel('PM2.5 in micrograms per cubic meter')
ax.legend(['Off','Fan Speed 1', 'Fan Speed 2', 'Fan Speed 3'],loc=1)
ax.margins(x=.01, y=0)
ax.text(x = -7, y = 1650, s = 'Sqair Artificial Pollution Test', fontsize=30, weight='bold', alpha=.8)
ax.text(x = -7, y = 1550, s= 'Room of 10 square meters (30 cubic meters of air)', fontsize=20)
ax.text(x = -9, y = -200, s = '  Â©Mongolian Data Stories                                                                                                                                              ', fontsize = 14, color = '#f0f0f0', backgroundcolor = 'dimgrey')
plt.savefig('images/test.png', dpi = 300, bbox_inches="tight")
```
