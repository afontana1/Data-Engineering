```python
def createDF(dat):
    data = dat.find("table")

    # getheaders
    head = []
    for th in data.findAll("th"):
        if th.text != "":
            head.append(th.text)

    tds = data.findAll("td")
    fin = []
    for t in range(len(tds)):
        chars = tds[t].text
        if chars == "":
            continue
        if len(chars.split()) > 1 or chars.isalpha():
            temp = "".join([val for sublist in chars.split() for val in sublist])
            out = []
            i = t + 1
            while tds[i].text.replace(".", "").isdigit():
                out.append(tds[i].text)
                i += 1
                if i == len(tds):
                    break
            fin.append({temp: out})

    df = []
    for d in fin:
        df.append(pd.DataFrame.from_dict(d, orient="columns").T)
    give = pd.concat(df)
    give.columns = head
    return give
```

```python
def matmul(mat1, mat2):
    """multiply two matrices"""
    numrows1, numcols1 = len(mat1), len(mat1[0])
    numrows2, numcols2 = len(mat2), len(mat2[0])
    if numcols1 != numrows2:
        return "mat1 cols do not match mat2 rows"
    outputmat = [[0 for i in range(numcols2)] for j in range(numrows1)]
    for i in range(len(mat1)):
        for j in range(len(mat2)):
            for k in range(len(outputmat[0])):
                outputmat[i][k] += mat1[i][j] * mat2[j][k]
    return outputmat


def diagonal(n):
    """returns the identity mat of size NxN"""
    return [[1 if i == j else 0 for i in range(n)] for j in range(n)]


def createMat(n, m, prepend1s=False, rands = False):
    """create mat of n rows and m cols"""
    if prepend1s:
        return [[1 if i == 0 else 0 for i in range(m + 1)] for j in range(n)]
    if rands:
        import random
        return [[random.randint(0,50) for i in range(m)] for j in range(n)]


def AddorSubtractor(mat1, mat2,operator):
    if (len(mat1) != len(mat2)) or (len(mat1[0]) != len(mat2[0])):
        return "need same number of rows and cols"
    outputmat = [[0 for i in range(len(mat1[0]))] for j in range(len(mat1))]
    
    for i in range(len(mat1)):
        for j in range(len(mat1[0])):
            if operator == "+":
                outputmat[i][j] = mat1[i][j] + mat2[i][j]
            if operator == "-":
                outputmat[i][j] = mat1[i][j] - mat2[i][j]                
                
    return outputmat

def MultorDiv(mat,val,operator):
    '''multiply or divide by a scalar'''
    if operator == "/":
        return [[mat[j][i]//val for i in range(len(mat[0]))] for j in range(len(mat))]
    return [[mat[j][i]*val for i in range(len(mat[0]))] for j in range(len(mat))]
    

def transpose(mat):
    '''convert nXm matrix to mXn'''
    outputmat = [[0 for i in range(len(mat))] for j in range(len(mat[0]))]
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            outputmat[j][i] = mat[i][j]
    return outputmat

def arePermsEqualParity(perm0, perm1):
    """Check if 2 permutations are of equal parity.

    Assume that both permutation lists are of equal length
    and have the same elements. No need to check for these
    conditions.
    """

    transCount = 0
    for loc in range(len(perm0) - 1):                         # Do (len - 1) transpositions
        if perm0[loc] != perm1[loc]:
            sloc = perm1.index(perm0[loc])                    # Find position in perm1
            perm1[loc], perm1[sloc] = perm1[sloc], perm1[loc] # Swap in perm1
            transCount += 1

    # Even number of transpositions means equal parity
    if (transCount % 2) == 0:
        return True
    else:
        return False
    
def BigDeterminant(mat):
    '''using Leibniz formula, very inefficient, factorial time'''
    if len(mat[0]) != len(mat):
        return 'need a square mat'
    det = 0
    firstRow = range(len(mat[0]))
    perms = list(itertools.permutations(firstRow))
    for i in range(len(perms)):
        permutations = list(perms[i])
        tups = list(zip(firstRow,permutations))
        something = [mat[tup[0]][tup[1]] for tup in tups]
        prod = reduce(lambda x,y:x*y,something)
        if arePermsEqualParity(firstRow,permutations):
            det+=prod
        else:
            det-=prod

    return det

def twoBytwo(mat):
    '''returns determinant of 2X2 mat, helper function'''
    denom = mat[0][0]*mat[1][1] - mat[0][1]*mat[1][0]
    return denom

def getSubMat(mat):
    '''get the sub matrices for a matrix'''
    for i in range(len(mat)):
        temp = []
        for j in range(len(mat)):
            temp.append(mat[j][:i] + mat[j][i+1:])
        yield temp[1:]

def determinant(mat,det = 0):
    '''using recursive definition of nXn matrix'''
    if len(mat) == 2:
        return twoBytwo(mat)
    if len(mat) > 2:
        for i,j in enumerate(getSubMat(mat)):
            if i%2 == 0:
                det+=mat[0][i]*determinant(j)
            if i%2 == 1:
                det-=mat[0][i]*determinant(j)
    return det

def invertMat(mat):
    if len(mat) != len(mat[0]):
        return "need a square matrix dude"
    return np.linalg.inv(mat)
```
